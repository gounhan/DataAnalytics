```python
import matplotlib.pyplot as plt

# Data setup
fruits = ['Apples', 'Bananas', 'Grapes', 'Oranges']
percentages = [30, 25, 20, 25]  # Preference percentages for each fruit

# Create pie chart
plt.pie(percentages, labels=fruits, autopct='%1.1f%%', startangle=140)

# Add title to the chart
plt.title('Favorite Fruits of 100 Students')

# Ensure the pie chart is a perfect circle
plt.axis('equal')

# Display the chart
plt.show()
```
